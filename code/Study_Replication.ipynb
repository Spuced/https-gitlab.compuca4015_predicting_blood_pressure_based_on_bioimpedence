> ##### **Default Approach**

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.model_selection import cross_validate
from sklearn.impute import SimpleImputer
import numpy as np
from sklearn.model_selection import LeavePGroupsOut
```

Based on the github and the paper the default approach is to train an AdaBoost regressor on the HGCP beats in the f2_ma10_mn1_cn1_all.csv file.

```python
training_df = df[df['setup_name'].str.contains('hgcp')]
training_df
```

They impute NaNs using the mean values of the whole dataset.

Ideally imputation is done separately on each cross-validation fold to avoid data leakage.

```python
def impute_nans(data):
    imp = SimpleImputer(missing_values=np.nan, strategy='mean')
    imp.fit(data)
    data_imp = imp.transform(data)
    return data_imp
```

DBP model

```python
X = training_df.drop(columns=['Sample', 'subject_id', 'setup_n', 'trial_n', 'fc', 'DBP', 'SBP', 'setup_name', 'exp_name', 'Sample_c', 'exp_setup_name'])
y = training_df['DBP']

X = impute_nans(X)

# Initialise the AdaBoostRegressor
# They don't specify the model parameters so this will only be a rough replication
ada_boost = AdaBoostRegressor(random_state=1234, n_estimators=8)

# Perform 10-fold cross-validation with multiple scoring metrics
scoring_metrics = ['neg_mean_squared_error', 'r2']
scores = cross_validate(ada_boost, X, y, cv=10, scoring=scoring_metrics)

# Calculate RMSE from MSE scores
rmse_scores = np.sqrt(-scores['test_neg_mean_squared_error'])

# MSE scores
mse_scores = -scores['test_neg_mean_squared_error']
print(f"MSE scores for each fold: {mse_scores}")
print(f"Mean cross-validation MSE: {np.mean(mse_scores)} +- {np.std(mse_scores)}")

# RMSE scores
print(f"RMSE scores for each fold: {rmse_scores}")
print(f"Mean cross-validation RMSE: {np.mean(rmse_scores)} +- {np.std(rmse_scores)}")

# R^2 scores
r2_scores = scores['test_r2']
print(f"R^2 scores for each fold: {r2_scores}")
print(f"Mean cross-validation R^2: {np.mean(r2_scores)} +- {np.std(r2_scores)}")
```

> ##### **Leave One Subject Out**

KFold can be misleading in cases like this, as data leaks between the folds, causing the model's performance to be overestimated.

LeavePGroups out can train the model on 5 subjects, and evaluate on the 6th, giving a better indication of the model's ability to generalise.

DBP

```python
X = training_df.drop(columns=['Sample', 'subject_id', 'setup_n', 'trial_n', 'fc', 'DBP', 'SBP', 'setup_name', 'exp_name', 'Sample_c', 'exp_setup_name'])
y = training_df['DBP']
groups = training_df['subject_id']

X = impute_nans(X)

# Initialise the AdaBoostRegressor
# They don't specify the model parameters so this will only be a rough replication
ada_boost = AdaBoostRegressor(random_state=1234)

# Leave-P-Groups-Out cross-validator
lpgo = LeavePGroupsOut(n_groups=1)

# Perform 10-fold cross-validation with multiple scoring metrics
scoring_metrics = ['neg_mean_squared_error', 'r2']
scores = cross_validate(ada_boost, X, y, groups=groups, cv=lpgo, scoring=scoring_metrics)

# Calculate RMSE from MSE scores
rmse_scores = np.sqrt(-scores['test_neg_mean_squared_error'])

# MSE scores
mse_scores = -scores['test_neg_mean_squared_error']
print(f"MSE scores for each fold: {mse_scores}")
print(f"Mean cross-validation MSE: {np.mean(mse_scores)} +- {np.std(mse_scores)}")

# RMSE scores
print(f"RMSE scores for each fold: {rmse_scores}")
print(f"Mean cross-validation RMSE: {np.mean(rmse_scores)} +- {np.std(rmse_scores)}")

# R^2 scores
r2_scores = scores['test_r2']
print(f"R^2 scores for each fold: {r2_scores}")
print(f"Mean cross-validation R^2: {np.mean(r2_scores)} +- {np.std(r2_scores)}")
```
